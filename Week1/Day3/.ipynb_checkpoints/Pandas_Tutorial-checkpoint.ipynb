### ILOC
In Pandas, `.iloc` is an integer-location based accessor used to select and modify data using numerical indices (0-indexed). It takes row and column positions separated by a comma: 
```python
df.iloc[#row_positions, #column_positions]
```

Unlike `.loc`, `.iloc` slices follow standard Python rules and exclude the stop index.

````python
coffee.iloc[:, 1:3]
````

### Practice Questions

#### 1. Find athletes born in cities that starts with a vowel

````python
bios[bios['born_city'].str.contains(r'^[AEIOUaeiou]', na = False)]
````

#### 2. Find athletes with names that contain exactly two vowels

````python
bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*[AEIOUaeiou]$', na = False)]
````

#### 3. Find athletes with names that have repeated consecutive letters (e.g., "Aaron", "Emmett")

````python
bios[bios['name'].str.contains(r'(.)\1', na = False)]
````

#### 4. Find athletes with names ending in `son` or `sen`

````python
bios[bios['name'].str.contains(r'son$|sen$', case = False, na = False)]
````

#### 5. Find athletes born in a year starting with `19`

````python
bios[bios['born_date'].str.contains(r'^19', na = False)]
````

#### 6. Find atheletes with names that do not contain any vowels

````python
bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*$', na = False)]
````

#### 7. Find athletes whose names contain a hyphen `-` or an apostrophe `'`

````python
bios[bios['name'].str.contains(r"[-']", na = False)]
````

#### 8. Find athletes with names that start and end with the same letter

````python
bios[bios['name'].str.lower().str.contains(r'^(.).*\1$', na = False, regex=True)]
````

###### A Quick Visual Trace:
If the engine checks the name "anna" (lowercased):
- ^ -> Starts at the beginning.
- (.) -> Saves a into Group 1.
- .* -> Skips over the middle letters (nn).
- \1 -> Sees the letter a, checks the memory bank, and confirms it matches Group 1.
- $ -> Checks if the string ends right here.

It does! The name is successfully kept.

#### 9. Find athletes with a `born_city` that has exactly 7 characters

````python
bios[bios['born_city'].str.contains(r'^.{7}$', na = False)]
````

Without anchors `r'.{7}'`: It will match any city that has at least 7 characters.
For example, "Chicago" (7) matches, but "Los Angeles" (11) also matches because pandas can find a 7-character chunk inside it.

We can also use 
```python
# Finds cities where the character length is exactly 7
bios[bios['born_city'].str.len() == 7]
```

````python
bios[bios['born_city'].str.len() == 7]
````

#### 10. Find athletes with names containing three or more vowels

````python
# grave mistake :
# ==> bios[~bios['name'].str.contains(r'^[^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*[AEIOUaeiou]$', na = False)] <==
# using this inverted logic of finding names with exactly 2 vowels not satisfactory in this case cz it will return names with 0 and 1 vowel too.
# stupid of me lol
````

````python
bios[bios['name'].str.contains(r'([AEIOUaeiou].*){3,}', na = False)]
````

Why this matters in the above vowel code `r'([AEIOUaeiou].*){3,}'`:

Because we put `.*` right after the vowel, the code says:
- Find a vowel.
- `.*` -> Allow any amount of consonants, spaces, or other letters to sit after it.
- Repeat that loop at least `3` times.

Without that `.*`, your code would look like `r'([AEIOUaeiou]){3,}'`, which would only match names where three vowels are jammed right next to each other with no consonants in between (like the "uai" in Luai).

````python
# Count all vowels (case-insensitive) and check if the count is 3 or more
bios[bios['name'].str.count(r'(?i)[aeiou]') >= 3]
````

#### Merge vs Concat
- Merge happens within same dataframe by merging columns
- Concat stacks one dataframe on top of another dataframe to make a combined dataframe


| Feature | `pd.concat()` | `pd.merge()` |
| :--- | :--- | :--- |
| **Primary Action** | Stacking or appending | Combining by key matching |
| **How it aligns** | By Index or Column Labels | By Column Values (Keys) |
| **Direction** | Vertically (Rows) or Horizontally (Cols) | Horizontally (Columns) only |
| **Best used for** | Combining identical files/logs | Combining lookup tables or relational data |



---

When DataFrames have mismatched columns, `pd.concat([a, b])` aligns matching columns vertically and automatically fills missing values with NaN. The overall behavior depends on the join parameter, which is set to outer by default.

1. Outer Join (Default)
```python
import pandas as pd

a = pd.DataFrame({'X': [1, 2], 'Y': [3, 4]})
b = pd.DataFrame({'X': [5, 6], 'Z': [7, 8]})

# Result will have columns X, Y, Z. 
# Y will be NaN for b's rows, Z will be NaN for a's rows.
result = pd.concat([a, b]) 
```
2. Inner Join (Common column only)
```python
# Result will only have column X
result_inner = pd.concat([a, b], join='inner')
```

#### Pro-Tips to Handle Misaligned Columns:

1. Reset the Index: Use ignore_index=True to avoid repeating structural index numbers (like having two `0` labels) in your new output.
2. Watch Out for Typos: If columns like `Age` and `age` do not match due to letter casing or trailing spaces, Pandas treats them as completely separate variables. Clean your columns beforehand using df.columns = `df.columns.str.strip().str.lower()`.
3. Fill Empty Fields Later: If you use the default outer join, clean up your missing data afterward by applying `.fillna(0)` or `.fillna("Missing")`.

You are seeing all NaN values because coffee is a DataFrame, but your filtering syntax is meant for a Series (a single column).When you pass a full DataFrame boolean mask into a DataFrame (`coffee[coffee.isna()]`), pandas tries to align the dimensions. Instead of dropping the valid rows, it replaces every valid value with `NaN` and keeps the rows intact.

Here are the standard ways to fix this, depending on what you want to see:
1. View rows where ANY column is missing:
If you want to see the actual data for rows that have at least one missing value, use `.any(axis=1)`
```python
coffee[coffee.isna().any(axis=1)]
```

2. View rows where a SPECIFIC column is missing:
If you only care about missing values in one column (for example, Units Sold), filter by that column's mask.
```python
coffee[coffee['Units Sold'].isna()]
```

3. View rows where ALL columns are missing:
If you suspect these rows are completely empty and want to verify, use `.all(axis=1)`
```python
coffee[coffee.isna().all(axis=1)]
```

````python
coffee[coffee.isna().any(axis=1)]
````

````python
coffee[coffee.isna().all(axis=1)]
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'bios' and already loaded
# First, filter out rows where the height_cm data is missing
bios_filtered = bios.dropna(subset=['height_cm'])

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(bios_filtered['height_cm'], bins=30, color='blue', edgecolor='black')

plt.title('Distribution of Athlete Heights in Olympics')
plt.xlabel('Height (in cm)')
plt.ylabel('Number of Athletes')
plt.grid(True)

# Using a logarithmic scale for the y-axis if the data spread is wide
plt.yscale('log')

plt.show()
````
